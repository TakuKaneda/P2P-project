### Peer to peer streaming capacity

We implement the method proposed by ... to construct the multitree to obtain the most efficient streaming rate in a P2P network.

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_edge(1,2)
K_5=nx.complete_graph(5)
nx.draw(K_5)
plt.show()
```
